```python
#Stage 2: Follow category pages
import requests
from bs4 import BeautifulSoup

# Stage 2: Follow the 4 main categories and get their sub-links

base_url = "https://www.unsw.edu.au"
start_url = "https://www.unsw.edu.au/study"

# Define the 4 categories from your screenshot
categories_to_crawl = [
    "Undergraduate",
    "Postgraduate",
    "International",
    "Professional Development"
]

# Step 1: Get the main study page
response = requests.get(start_url)
soup = BeautifulSoup(response.content, "html.parser")

# Step 2: Find links for just these categories
category_links = {}

for link in soup.find_all("a"):
    text = link.text.strip()
    if text in categories_to_crawl:
        href = link.get("href")
        if href:
            # Make absolute URL if it's relative
            if href.startswith("/"):
                href = base_url + href
            category_links[text] = href

# Step 3: Visit each category and extract sub-links
for category, url in category_links.items():
    print(f"\n===== {category} =====")
    resp = requests.get(url)
    cat_soup = BeautifulSoup(resp.content, "html.parser")

    # Find all submenu links on that page
    for sub_link in cat_soup.find_all("a"):
        sub_text = sub_link.text.strip()
        sub_href = sub_link.get("href")

        if sub_text and sub_href:
            if sub_href.startswith("/"):
                sub_href = base_url + sub_href
            # Print sub-links
            print(f"{sub_text} -> {sub_href}")
```
